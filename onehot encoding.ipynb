### One Hot Encoder

#### Using sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import pandas as pd
```

```python
data2=pd.read_csv("Company_Data.csv",index_col=0)
```

```python
data2
```

```python
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
# passing bridge-types-cat column (label encoded values of bridge_types)
enc_df = pd.DataFrame(enc.fit_transform(data2[['ShelveLoc']]).toarray())
```
